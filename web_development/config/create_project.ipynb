## Create a react project for web.
### Install specific version of Tailwind due to install error.


```python
# enter into web folder
cd web_dic
# use Vite to create a React TypeScript project
npm create vite@latest . -- --template react-ts
# install dependencies
npm install
# install Tailwind CSS and its dependencies
npm install tailwindcss@latest postcss@latest autoprefixer@latest
# if install failed, try the stable version of Tailwind CSS
npm install tailwindcss@3.4.3 postcss@8.4.38 autoprefixer@10.4.19
# initialize Tailwind CSS configuration
npx tailwindcss init -p
# configure tailwind.config.js
content:[
  "./index.html",
  "./src/**/*.{js,ts,jsx,tsx}",
  "./src/**/*.{css,scss}"
]
# add Tailwind directives to src/index.css
@tailwind base;
@tailwind components;
@tailwind utilities;
# before starting the server, you can add this to make sure the server will open automatically
# add the configuration in vite.config.ts
server: {
  open: true
}
# start the development server
npm run dev
```
